```python
import pandas as pd
import numpy as np
```

Creo los dataframes

```python
df_Amazon = pd.read_csv('Datasets/amazon_prime_titles.csv')
df_Disney = pd.read_csv('Datasets/disney_plus_titles.csv')
df_Hulu = pd.read_csv('Datasets/hulu_titles.csv')
df_Netflix = pd.read_json('Datasets/netflix_titles.json')
```

elimino duplicados

```python
df_Amazon = df_Amazon.drop_duplicates()
df_Disney = df_Disney.drop_duplicates()
df_Hulu = df_Hulu.drop_duplicates()
df_Netflix = df_Netflix.drop_duplicates()
```

creo una nueva columna en todos los dataframes con el nombre de la plataforma, esto ayudara mas adelante

```python
df_Amazon = df_Amazon.assign(Plataform = 'Amazon Prime')
df_Amazon
```

```python
df_Disney = df_Disney.assign(Plataform = 'Disney Plus')
df_Disney
```

```python
df_Hulu = df_Hulu.assign(Plataform = 'Hulu')
df_Hulu
```

```python
df_Netflix = df_Netflix.assign(Plataform = 'Netflix')
df_Netflix
```

Uno los df

```python
df_Total = pd.concat([df_Amazon,df_Disney,df_Hulu,df_Netflix])
df_Total
```

elimino la columna show_id, ya que hay ids que se repiten por pertenecer a diferentes plataformas

```python
df_Total = df_Total.drop(['show_id'], axis = 1) 
df_Total
```

Elimino posibles duplicados

```python
df_Total = df_Total.drop_duplicates()
df_Total
#no habia duplicados
```

Creo una nueva columna id

```python
df_Total['Id'] = np.arange(1,len(df_Total)+1)
df_Total
```

cambio la posicion del id

```python
Id = df_Total.pop('Id')
df_Total.insert(0,'Id',Id)
```
